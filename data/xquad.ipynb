```python
from datasets import load_dataset, Dataset
import datasets
import pandas as pd
import random
import numpy as np
```

```python
# Load dataset
xquad_ca = load_dataset("projecte-aina/xquad-ca")
xquad_en = load_dataset("xquad", "xquad.en")
```

```python
id_to_drop = set(xquad_en['validation']['id']).difference(set(xquad_ca['test']['id'])).pop()
id_to_drop
```

```python
xquad_en = xquad_en['validation'].filter(lambda example: example['id'] != id_to_drop)
xquad_en
```

```python
xquad_en.to_pandas()
```

```python
xquad_ca["test"].to_pandas()
```

```python
df_ca = xquad_ca['test'].to_pandas()
df_en = xquad_en.to_pandas()
df_ca["answers"] = df_ca["answers"].map(lambda x: x[0])
```

```python
df_en
```

```python
# Parameters
seed = 42
few_shot_count = 3
np.random.seed(seed)


# Filter the DataFrame based on the contexts that appear more than few_shot_count times
context_counts = df_ca.groupby('context')['id'].count()
filter_min_few_shot = context_counts > few_shot_count
filtered_df = df_ca[df_ca['context'].isin(context_counts[filter_min_few_shot].index)]

dataset_ca = Dataset.from_pandas(df_ca)

# Tracker for samples
context_samples = {}

entries_ca = []
entries_en = []
for context, df_ca_group in df_ca.groupby('context'):
    for _id in df_ca_group['id']:
        few_shot_ids = df_ca.query(f'id != "{_id}"').sample(3)['id'].tolist()

        # Catalan
        samples = df_ca.query(f'id in @few_shot_ids')
        prompt_ca = f"{samples.iloc[0]['context']}" + "\n----\n"
        prompt_ca += "\n----\n".join(
            f"Pregunta: {sample['question']}\nResposta: {sample['answers']['text']}"
            for _, sample in samples.iterrows()
        )
        question_ca = df_ca.query(f'id == "{_id}"').iloc[0]
        prompt_ca += "\n----\n" + f"Pregunta: {question_ca['question']}\nResposta: "
        entries_ca.append({
            'context': context,
            'prompt': prompt_ca,
            'answer': question_ca['answers']['text']
        })

        # English
        samples = df_en.query(f'id in @few_shot_ids')
        prompt_en = f"{samples.iloc[0]['context']}" + "\n----\n"
        prompt_en += "\n----\n".join(
            f"Question: {sample['question']}\nResponse: {sample['answers']['text']}"
            for _, sample in samples.iterrows()
        )
        question_en = df_en.query(f'id == "{_id}"').iloc[0]
        prompt_en += "\n----\n" + f"Question: {question_en['question']}\nResponse: "
        entries_en.append({
            'context': samples.iloc[0]["context"],
            'prompt': prompt_en,
            'answer': question_en['answers']['text'][0]
        })
```

```python
xquad_df_ca = pd.DataFrame(entries_ca)
xquad_df_ca
```

```python
xquad_df_en = pd.DataFrame(entries_en)
xquad_df_en
```

```python
xquad_df_ca.to_csv('xquad_ca.csv', index=False)
xquad_df_en.to_csv('xquad_en.csv', index=False)
```
